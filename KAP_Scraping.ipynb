```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import StaleElementReferenceException
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
import pandas as pd
import random
import time
import random
import os
```

```python
# Set up Edge options
ua = UserAgent()
user_agent = ua.random
edge_options = Options()  # Optional: run Edge in headless mode
edge_options.add_argument(f"user-agent={user_agent}")

# Set up the Edge WebDriver service
service = Service()

# Create a new WebDriver instance for Edge
driver = webdriver.Edge(service=service, options=edge_options)

# Navigate to a webpage
driver.get('https://publikasi.ojk.go.id/Portal')
driver.implicitly_wait(10)
```

```python
tab_ap_button_xpath = "Akuntan Publik"
tab_ap = WebDriverWait(driver, 20).until(
    EC.presence_of_element_located((By.LINK_TEXT, tab_ap_button_xpath))
)
tab_ap.click()
```

```python
condition = True
data = []

while condition:
    time.sleep(2)
    akuntan_publik_xpath = "(//tbody)[2]/tr"
    akuntan_publik = driver.find_elements(By.XPATH, akuntan_publik_xpath)

    for akuntan in akuntan_publik:
        akuntan_data = {
                'Nama': akuntan.find_element(By.XPATH, 'td[1]').text,
                'Link': akuntan.find_element(By.XPATH, 'td[1]/a').get_attribute('href'),
                'KAP' : akuntan.find_element(By.XPATH, 'td[2]').text,
                'Jabatan' : akuntan.find_element(By.XPATH, 'td[3]').text,
                'Status' : akuntan.find_element(By.XPATH, 'td[9]').text
            }
        data.append(akuntan_data)
        print(akuntan_data)

    try:
        
        driver.find_element(By.LINK_TEXT, "Next").click()

        if akuntan_data['Nama'] == "Zulpan":
            condition = False
    
    except:
        condition = False
```

# Ambil Data Masing Masing Orang

```python
data_orang = data.copy()
data_orang = pd.DataFrame(data_orang)
data_orang
```

```python
data_akuntan_final = []
```

```python
for index,row in data_orang.iterrows():
    try:
        time.sleep(3)
        driver.get(row['Link'])
        data_akuntan = {
            'Nama' : row['Nama'],
            'KAP' : row['KAP'],
            'Jabatan' : row['Jabatan'],
            'Status' : row['Status'],
            'Telepon' : driver.find_element(By.XPATH, "(//table)[1]/tbody/tr[3]/td[3]").get_attribute("textContent"),
            'Faximail' : driver.find_element(By.XPATH, "(//table)[1]/tbody/tr[5]/td[3]").get_attribute("textContent"),
            'Email' : driver.find_element(By.XPATH, "(//table)[1]/tbody/tr[7]/td[3]").get_attribute("textContent"),
        }
        data_akuntan_final.append(data_akuntan)
        print(data_akuntan)
    except:
        pass
```

```python
data_akuntan_final = pd.DataFrame(data_akuntan_final)
data_akuntan_final.to_csv("data_akuntan.csv", index=False)
```
